1) Считайте датасет из файла Реестр_выполненных_рейсов_за_период.xlsx

```python
import pandas as pd
df = pd.read_excel('/content/Реестр_выполненных_рейсов_за_период.xlsx')
df.head()
```

2) Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д.

```python
print("=== Информация о датасете ===")
df.info()

print("\n=== Число пропусков в каждом столбце ===")
display(df.isnull().sum())

print("\n=== Основные статистические характеристики числовых столбцов ===")
display(df.describe())

print("\n=== Размер датасета (строк, столбцов) ===")
display(df.shape)

print("\n=== Названия столбцов ===")
display(df.columns.tolist())
```

3) Посчитайте процент выполненных рейсов по каждому из аэропортов (Код АП), выведите топ-10 результатов - 1 балл

hint: используйте методы value_counts, head, для расчета долей пригодится параметр normalize

```python
percent_by_airport = df['Код АП'].value_counts(normalize=True) * 100

display(percent_by_airport.head(10))
```

4) Посчитайте то же самое, что и в пункте выше, только для Domestic рейсов - 1 балл

hint: перед расчетами выполните фильтрацию по нужному столбцу

```python
domestic_df = df[df['Направление'] == 'Domestic']

percent_by_airport_domestic = domestic_df['Код АП'].value_counts(normalize=True) * 100

display(percent_by_airport_domestic.head(10))
```

5) Выведите самый популярный маршрут по каждому из направлений (Domestic/International)

hint: используйте методы group, agg, pd.Series.mode или фильтрацию и value_counts

```python
# Самый популярный аэропорт среди Domestic
domestic_df = df[df['Направление'] == 'Domestic']
most_common_airport_domestic = domestic_df['Код АП'].value_counts().idxmax()
print(f"Самый популярный аэропорт для Domestic: {most_common_airport_domestic}")
```

```python
# Самый популярный аэропорт среди International
international_df = df[df['Направление'] == 'International']
most_common_airport_international = international_df['Код АП'].value_counts().idxmax()
print(f"Самый популярный аэропорт для International: {most_common_airport_international}")
```

6) Выведите часть таблицы с рейсами категории C и маршрутом САЛЕХАРД

```python
filtered_df = df[(df['Катег'] == 'C') & (df['Маршрут'] == 'САЛЕХАРД')]
display(filtered_df)
```

7) Выведите на экран результат из пункта выше с сортировкой по убыванию Baggage Weight

```python
sorted_df = filtered_df.sort_values(by='Baggage Weight', ascending=False)
display(sorted_df)
```

8) Выведите среднюю МВМ и Вместимость с группировкой по направлению и типу ВС - 2 балла

hint: используйте методы group и mean, группировать и считать статистики можно сразу по нескольким полям

```python
result = df.groupby(['Направление', 'Тип ВС'])[['МВМ, тонн', 'Вместимость']].mean()
display(result)
```
