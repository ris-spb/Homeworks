1) Считайте датасет из файла Реестр_выполненных_рейсов_за_период.xlsx

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_excel('/content/Реестр_выполненных_рейсов_за_период.xlsx')
df.head()
```

2) Визуализируйте распределение значений признаков Вместимость, Обслужено пассажиров, Baggage Weight, МВМ, тонн.

```python
columns = ['Вместимость', 'Обслужено пассажиров', 'Baggage Weight', 'МВМ, тонн']

for col in columns:
    plt.figure(figsize=(8, 4))
    df[col].dropna().hist(bins=50)
    plt.title(f'Распределение: {col}')
    plt.xlabel(col)
    plt.ylabel('Частота')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

3) Постройте график типа boxplot для столбца Baggage Weight

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df['Baggage Weight'].dropna(), vert=True, patch_artist=True)
plt.title('Boxplot: Baggage Weight')
plt.ylabel('Вес багажа')
plt.grid(True)
plt.show()
```

На боксплоте видно, что медиана веса багажа составляет ~500, а межквартильный размах — от ~300 до ~700. Есть выбросы в отрицательной области (вероятно ошибки), так и в положительной

4) Постройте график типа pie chart для переменной В/П, подпишите доли в процентах

```python
vp_counts = df['В/П'].dropna().value_counts(normalize=True) * 100

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(vp_counts, labels=vp_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение по переменной В/П')
plt.axis('equal')  # круглая диаграмма
plt.show()
```
